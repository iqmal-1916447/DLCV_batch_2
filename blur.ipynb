```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
images_shot_bgr = []
images_shot_rgb = []
for i in range(3):
  images_shot_bgr.append(cv2.imread('copy_of_photo_' + str(i) + '.jpg', cv2.IMREAD_COLOR))
  images_shot_rgb.insert(i, images_shot_bgr[i][:,:,::-1])
  plt.subplot(1,3,i+1)
  plt.imshow(images_shot_rgb[i])
```

```python
# draw rectangles

images_rect_annt = images_shot_rgb.copy()
for i in range(3):
  images_rect_annt[i] = images_rect_annt[i].astype(np.uint8)

cv2.rectangle(images_rect_annt[0], (270, 100), (500, 350), (0, 0, 255), thickness=5, lineType=cv2.LINE_8)
plt.subplot(1,3,1)
plt.imshow(images_rect_annt[0])

cv2.rectangle(images_rect_annt[1], (270, 95), (500, 350), (0, 0, 255), thickness=5, lineType=cv2.LINE_8)
plt.subplot(1,3,2)
plt.imshow(images_rect_annt[1])

cv2.rectangle(images_rect_annt[2], (270, 100), (500,350), (0, 0, 255), thickness=5, lineType=cv2.LINE_8)
plt.subplot(1,3,3)
plt.imshow(images_rect_annt[2])

```

```python
def get_kernel(image):
  h, w = image.shape[:2]
  kernel_width = (w // 7) | 1
  kernel_height = (h // 7) | 1
  return (kernel_width, kernel_height)
```

```python
faces = []

# face_0

faces.insert(0, images_rect_annt[0][100:350, 270:500]) 
faces[0] = cv2.GaussianBlur(faces[0], get_kernel(images_shot_rgb[0]), 0)
images_rect_annt[0][100:350, 270:500] = faces[0]


plt.imshow(images_rect_annt[0])
```

```python
# face_1

faces.insert(1, images_rect_annt[1][95:350, 270:500]) 
faces[1] = cv2.GaussianBlur(faces[1], get_kernel(images_shot_rgb[1]), 0)
images_rect_annt[1][95:350, 270:500] = faces[1]


plt.imshow(images_rect_annt[1])
```

```python
# face_2
faces.insert(2, images_rect_annt[2][100:350, 270:500])
faces[2] = cv2.GaussianBlur(faces[2], get_kernel(images_shot_rgb[2]), 0)
images_rect_annt[2][100:350, 270:500] = faces[2]

plt.imshow(images_rect_annt[2])
```
